```python
import os
```

```python

base_dir = "../../data/chest_xray"  # Adjust path if needed

for split in ["train", "val", "test"]:
    print(f"\nüìÅ {split.upper()}:")

    for label in ["NORMAL", "PNEUMONIA"]:
        folder = os.path.join(base_dir, split, label)
        count = len(os.listdir(folder))
        print(f"  {label}: {count} images")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random

def show_sample_images(label="PNEUMONIA", split="train", n=5):
    folder = os.path.join(base_dir, split, label)
    filenames = random.sample(os.listdir(folder), n)

    plt.figure(figsize=(15, 5))
    for i, file in enumerate(filenames):
        img_path = os.path.join(folder, file)
        img = mpimg.imread(img_path)

        plt.subplot(1, n, i+1)
        plt.imshow(img, cmap="gray")
        plt.title(f"{label}")
        plt.axis("off")

    plt.show()

show_sample_images("PNEUMONIA")
show_sample_images("NORMAL")
```
